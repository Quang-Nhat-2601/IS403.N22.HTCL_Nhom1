# Metrics


## btc

```python
model_name = [
    "arima",
    "cnn",
    "linear_regression",
    "lstm_att",
    "lstm",
    "random_forest",
    "rnn",
    "seq2seq",
    "xgboost",
    "sarimax"
]
token_id = 0
metrics_df_0 = pd.DataFrame()

for model in model_name:
  df = pd.read_excel(f"metrics/{tokens[token_id]}/{model}_{train_test_split[n_split]}.xlsx")
  df.drop(columns=["Unnamed: 0"], inplace=True)
  metrics_df_0 = pd.concat([metrics_df_0, df])

metrics_df_0["algo"] = model_name
metrics_df_0.set_index("algo", inplace=True)

metrics_df_0.fillna(99999, inplace=True)
metrics_df_0.sort_values("rmse_test", axis = 0, ascending = True,
                 inplace = True, na_position ='first')
metrics_df_0.to_excel(f"metrics/all_metrics_{tokens[token_id]}_sorted_{train_test_split[n_split]}.xlsx")
metrics_df_0
```
